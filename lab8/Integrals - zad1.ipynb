```julia
using QuadGK
using Polynomials
using Plots
```

```julia
function Lagrange(k)
    if k==0
        return Poly([1])
    elseif k==1
        return Poly([0,1])
    end
    k=k-1
    (2*k+1)/(k+1)*Poly([0,1])*Lagrange(k)-k/(k+1)*Lagrange(k-1)
end
```

```julia
Deg = 0:1:5
X=-1:0.02:1
Y=[[Lagrange(deg)(x) for x in X] for deg in Deg]
labels = [string("Lagrange(",deg,")") for deg in Deg]
```

```julia
plot(X,Y, title = "Wielomiany Lagrange", labels = labels)
```

```julia
gaussPlot = [
    plot(X, Y[i+1], title=string("Wielomian stopnia ",i)) for i in 1:4
]


d=[
    [0.0] for i in 1:4
]

for i in 1:4
    (xp,a)=gauss(Float64,i)
    yp = [0 for i in xp]
    gaussPlot[i] = plot(X, Y[i+1], title=string("Wielomian stopnia ",i))
    plot!(xp, yp, seriestype=:scatter, legend=false)
    d[i] = sort(roots(Lagrange(i)))-sort(xp)
end
```
