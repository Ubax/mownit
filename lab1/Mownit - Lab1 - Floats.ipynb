```julia
using Plots
x=Float32[]
y=Float32[]
for i=1:10000
    push!(x,i*1.0)
    push!(y,nextfloat(x[i]))
end
```

```julia
plot(x,y,title="Diff")
```

```julia
x=Float64[]
y=Float64[]
for i=1:1000
    push!(x,i*1.0)
    push!(y,nextfloat(x[i])-x[i])
end
```

```julia
plot(x,y,title="Diff")
```
